```python
#STEP 1: Load Dataset
import pandas as pd
df = pd.read_csv('store_dataset.csv', encoding='latin1')
df.head()
```

```python
#STEP 3: Data Cleaning
df.isnull().sum()
```

```python
#Convert Date Columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
#Create New Columns (Feature Engineering)
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
#STEP 4: Exploratory Data Analysis (EDA)
#Overall Business Metrics

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
#Sales by Region

import matplotlib.pyplot as plt

df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
#Sales by Category

df.groupby('Category')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Category")
plt.show()
```

```python
#Monthly Sales Trend

df.groupby('Month')['Sales'].sum().plot(kind='line')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
#Top 10 Products

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='barh')
plt.title("Top 10 Products")
plt.show()
```

```python
#Profit by Region

df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Region")
plt.show()
```

```python
#Correlation Analysis

import seaborn as sns

sns.heatmap(df[['Sales','Profit','Quantity','Discount']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
#STEP 5: Advanced Analysis (Optional but Powerful)
#Loss-making Products

loss_products = df[df['Profit'] < 0]
loss_products[['Product Name','Profit']].head()
```

```python
#High Discount Impact

df.groupby('Discount')['Profit'].mean()
```

```python
#STEP 6: Export Clean Data

df.to_csv("cleaned_sales_data.csv", index=False)
```
